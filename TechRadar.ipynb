```python
import requests

URL="https://www.techradar.com/news/archive/2021/03"

page=requests.get(URL)
print(page.content)

from bs4 import BeautifulSoup
soup=BeautifulSoup(page.content,'html.parser')
news_main=soup.find_all("ul",class_="day-list")
```

```python
TRnews23=news_main[1]
TRnews23.contents
```

```python
TRnews23links=TRnews23.find_all('a',href=True)

TRnews23links[0]
```

```python
TRnum_articles23=20

for c in range(0,TRnum_articles23):
    link = TRnews23links[c].get('href')
    article = requests.get(link)
    article_content = article.content
    soup_article = BeautifulSoup(article_content, 'html5lib')
    
    title=soup_article.find('h1').get_text()
    author=soup_article.find('span', style="white-space: nowrap").get_text()
    date=soup_article.find('time', itemprop="datePublished").get_text()
    body = soup_article.find_all(id="article-body")
    w = body[0].find_all('p')
    
    list_paragraphs = []
    for p in range(0, len(w)-1):
        paragraph = w[p].get_text()
        list_paragraphs.append(paragraph)
        fullarticle = "\n".join(list_paragraphs)

    #print the articles
    print(f'\nArticle #', c+1)
    print('-------------')
    print(f'{title}')
    print(f'  By:{author}  | {date}\n') 
    print(fullarticle)
```
